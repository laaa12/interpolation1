```python
# Интерполяционный полином Лагранжа
def lagrange(x, xs, ys):
    s = 0
    n = len(xs)
    for i in range(n):
        p = 1
        for j in range(n):
            if i == j:
                continue
            diff = (x - xs[j]) / (xs[i] - xs[j])
            if abs(diff) > 1e10:  # Ограничиваем слишком большие значения
                print(f"Пропуск множителя из-за переполнения при x = {x}")
                p = np.nan
                break
            p *= diff
        if not np.isnan(p):
            s += ys[i] * p
    return s

# Функция для применения Лагранжа на каждом интервале
def lagrange_piecewise(x_intervals, xs, ys):
    result_x = []
    result_y = []
    
    for interval in x_intervals:
        subset_indices = (xs >= interval[0]) & (xs <= interval[1])
        xs_subset = xs[subset_indices]
        ys_subset = ys[subset_indices]
        
        # Генерация точек интерполяции в этом интервале
        x_interp = np.linspace(interval[0], interval[1], 50) # Можно варьировать количество взятых точек
        y_interp = []
        
        for xi in x_interp:
            y_val = lagrange(xi, xs_subset, ys_subset)
            if not np.isnan(y_val):
                result_x.append(xi)
                result_y.append(y_val)
            else:
                print(f"Пропущено значение NaN для x = {xi}")
                
    return np.array(result_x), np.array(result_y)

# Определяем интервалы для интерполяции
num_intervals = 30 # Можно варьировать количество интервалов                                                                         
x_intervals = np.array_split(np.linspace(min(unique_x), max(unique_x), 100), num_intervals)

# Применение кусочно-линейного Лагранжа
valid_x, valid_y = lagrange_piecewise(x_intervals, unique_x, unique_y)

# Визуализация интерполяционного полинома Лагранжа по подмножествам
plt.figure(figsize=(12, 8))
plt.scatter(unique_x, unique_y, color='blue', label='Исходные точки', s=10)
plt.plot(valid_x, valid_y, color='orange', label='Полином Лагранжа (кусочно)')
plt.title('Интерполяционный полином Лагранжа по подмножествам')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1, 1)
plt.grid()
plt.legend()
plt.show()
```

```python
# настройки краевых условий
left_derivative = (unique_y[1] - unique_y[0]) / (unique_x[1] - unique_x[0])  # Приблизительная производная на левом конце
right_derivative = (unique_y[-1] - unique_y[-2]) / (unique_x[-1] - unique_x[-2])  # Приблизительная производная на правом конце

# Перепишите cubic_spline для учета правой производной
def cubic_spline(x, y, left_derivative, right_derivative):
    n = len(x) - 1
    h = np.diff(x)
    alpha = np.zeros(n)

    # Вычисляем значения alpha, за исключением первого и последнего уравнений
    for i in range(1, n):
        alpha[i] = (3 / h[i]) * (y[i + 1] - y[i]) - (3 / h[i - 1]) * (y[i] - y[i - 1])

    # Обновляем первое и последнее уравнение для учета производной
    alpha[0] = 3 * ((y[1] - y[0]) / h[0] - left_derivative)
    alpha[n-1] = 3 * (right_derivative - (y[n] - y[n-1]) / h[n-1])

    # Система уравнений
    l = np.zeros(n + 1)
    mu = np.zeros(n)
    z = np.zeros(n + 1)

    l[0] = 2 * h[0]
    mu[0] = 0.5
    z[0] = alpha[0] / l[0]

    for i in range(1, n):
        l[i] = 2 * (x[i + 1] - x[i - 1]) - h[i - 1] * mu[i - 1]
        mu[i] = h[i] / l[i]
        z[i] = (alpha[i] - h[i - 1] * z[i - 1]) / l[i]

    l[n] = h[n - 1] * (2 - mu[n - 1])
    z[n] = (alpha[n - 1] - h[n - 1] * z[n - 1]) / l[n]

    c = np.zeros(n + 1)
    b = np.zeros(n)
    d = np.zeros(n)

    c[n] = z[n]

    for j in range(n - 1, -1, -1):
        c[j] = z[j] - mu[j] * c[j + 1]
        b[j] = (y[j + 1] - y[j]) / h[j] - h[j] * (c[j + 1] + 2 * c[j]) / 3
        d[j] = (c[j + 1] - c[j]) / (3 * h[j])

    return b, c, d

# Вычисление коэффициентов с обновленными краевыми условиями
b, c, d = cubic_spline(unique_x, unique_y, left_derivative, right_derivative)

# Интерполяция с помощью кубического сплайна
def spline_interpolation(x, y, b, c, d, x_val):
    n = len(x) - 1
    for i in range(n):
        if x[i] <= x_val <= x[i + 1]:
            dx = x_val - x[i]
            return y[i] + b[i] * dx + c[i] * dx**2 + d[i] * dx**3
    return None

# Генерация точек для сплайна
x_interpolated = np.linspace(min(unique_x), max(unique_x), 130)
y_spline_interpolated = [spline_interpolation(unique_x, unique_y, b, c, d, xi) for xi in x_interpolated]

# Визуализация кубического сплайна
plt.figure(figsize=(12, 8))
plt.scatter(unique_x, unique_y, color='blue', label='Исходные точки', s=10)
plt.plot(x_interpolated, y_spline_interpolated, color='green', label='Кубический сплайн')
plt.title('Кубический сплайн интерполяции')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```
